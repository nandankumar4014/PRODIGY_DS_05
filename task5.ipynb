```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
file_path =  r"D:\DS_tasks\accidents.csv"
data = pd.read_csv(file_path)
data
```

```python
data.head(5)
```

```python
# Plotting the distribution of accident severity
sns.set(style="whitegrid")
plt.figure(figsize=(5, 4))
sns.countplot(data=data, x='Accident_severity', palette='viridis')
plt.title('Distribution of Accident Severity')
plt.xlabel('Accident Severity')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
# Plotting the distribution of weather conditions
plt.figure(figsize=(5,4))
sns.countplot(data=data, y='Weather_conditions', palette='viridis', order=data['Weather_conditions'].value_counts().index)
plt.title('Accidents by Weather Conditions')
plt.xlabel('Number of Accidents')
plt.ylabel('Weather Conditions')
plt.show()

# Plotting the distribution of road surface types
plt.figure(figsize=(5,4))
sns.countplot(data=data, y='Road_surface_type', palette='viridis', order=data['Road_surface_type'].value_counts().index)
plt.title('Accidents by Road Surface Type')
plt.xlabel('Number of Accidents')
plt.ylabel('Road Surface Type')
plt.show()

# Plotting the distribution of light conditions
plt.figure(figsize=(5,4))
sns.countplot(data=data, y='Light_conditions', palette='viridis', order=data['Light_conditions'].value_counts().index)
plt.title('Accidents by Light Conditions')
plt.xlabel('Number of Accidents')
plt.ylabel('Light Conditions')
plt.show()
```

```python
# Analyzing the intersection of key factors

# Heatmap of accidents by weather conditions and road surface type
pivot_weather_surface = data.pivot_table(index='Weather_conditions', columns='Road_surface_type', aggfunc='size', fill_value=0)
plt.figure(figsize=(8, 4))
sns.heatmap(pivot_weather_surface, cmap='viridis', annot=True, fmt='d')
plt.title('Accidents by Weather Conditions and Road Surface Type')
plt.xlabel('Road Surface Type')
plt.ylabel('Weather Conditions')
plt.show()

```

```python
# Heatmap of accidents by light conditions and junction types
pivot_light_junction = data.pivot_table(index='Light_conditions', columns='Types_of_Junction', aggfunc='size', fill_value=0)
plt.figure(figsize=(8,4))
sns.heatmap(pivot_light_junction, cmap='viridis', annot=True, fmt='d')
plt.title('Accidents by Light Conditions and Junction Types')
plt.xlabel('Types of Junction')
plt.ylabel('Light Conditions')
plt.show()
```

```python
# Heatmap of accidents by accident severity and weather conditions
pivot_severity_weather = data.pivot_table(index='Accident_severity', columns='Weather_conditions', aggfunc='size', fill_value=0)
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_severity_weather, cmap='viridis', annot=True, fmt='d')
plt.title('Accident Severity by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Accident Severity')
plt.show()
```
